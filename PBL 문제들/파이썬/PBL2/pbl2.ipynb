```python
import pandas as pd
#csv 파일에서 ip 만 뽑아오기
df = pd.read_csv('wireshark_mock_realistic_data.csv')

#print(df)
source_destination = df[['Source', 'Destination']]

Src_list = []
Dst_list = []
count = 0

for line in source_destination['Source']:
    Src_list.append(line)

for line in source_destination['Destination']:
    Dst_list.append(line)

Ip_list = Src_list + Dst_list

no_overlap_ip = set(Ip_list)

data = {
    'ip' : [],
    'count' : []
}

ip_df = pd.DataFrame(data)

for ip in no_overlap_ip :
    count = Ip_list.count(ip)

    new_data = {
    'ip' : [f'{ip}'],
    'count' : [count]
    }

    df2 = pd.DataFrame(new_data)

    ip_df = pd.concat([ip_df, df2], ignore_index=True)

sorted_df = ip_df.sort_values(by='count', ascending=False)

with open('top3_ip.csv', 'w', encoding='utf8') as f:
    f.write('가장 많이 등장한 상위 3개의 IP 주소를 표시합니다 : \n')
    for ip, count in zip(sorted_df['ip'][:3], sorted_df['count'][:3]):
        f.write(f'IP : {ip} 빈도수 : {count} \n')

print('가장 많이 등장한 상위 3개의 IP 주소를 표시합니다 : ')
print(sorted_df[:3])
```
